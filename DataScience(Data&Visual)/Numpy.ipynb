```python
#identity
np.identity(3)
```

```python
#copy pattern
print(a)
print(a.shape)
print("")

print(np.full_like(a.shape, 4))
print("")
print(np.full_like(a, 4))
```

```python
#random decimal 0-1
np.random.rand(4,2,3)
```

```python
# random inetger value
np.random.randint(0,100,size=(3,3))
```

```python
#Repeat
arr=np.array([[1,2,3]])
r1 = np.repeat(arr,3,axis=1)
print(r1,"\n")

r2= np.repeat(arr,3,axis=0)
print(r2)
```

```python
output = np.ones((5,5))
print(output)

z=np.zeros((3,3))
z[1,1]=9
print(z)

output[1:-1,1:-1] = z
print(output)
```

```python
a=np.array([1,2,3])
b=a                 #連立
c=np.array([1,2,3])
d=c.copy()         #純copy content

b[0]=100
print(a)
print(b)
c[0]=100
print(c)
print(d)
```

# Organize an array

```python
before = np.array([[1,2,3,4],[5,6,7,8]])

after= before.reshape((2,2,2)) #2,2,2 is equal to 8

after
```

```python
#Vertically/Horizontally stacking vectors
v1 = np.array([1,2,3,4])
v2= np.array([5,6,7,8])

print(np.vstack([v1,v2,v1,v2]))
print(np.hstack([v1,v2,v1,v2]))
```
